```python
#Loading Mall Customer Segmentation Dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# load your dataset
df = pd.read_csv("Mall_Customers.csv")
```

```python
#Displaying first 5 rows
df.head()
```

```python
#Displaying Shape of the Dataset
print("Shape:", df.shape)
```

```python
#Displaying Columns of the Dataset
print("Columns:", df.columns.tolist())
```

```python
#Histogram with 15 bins
plt.figure()
plt.hist(df["Age"], bins=15)

plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

```python
#Barchart with Categorial values
gender_counts = df["Gender"].value_counts()

plt.figure()
gender_counts.plot(kind="bar", color=["skyblue","salmon"])

plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.show()
```

```python
#Box Plot for Outlier Detection
plt.figure()
sns.boxplot(x=df["Annual Income (k$)"])
plt.title("Box Plot of Annual Income")

plt.show()
```

```python
#Quartile interpretation
q1 = df["Annual Income (k$)"].quantile(0.25)
q2 = df["Annual Income (k$)"].quantile(0.50)
q3 = df["Annual Income (k$)"].quantile(0.75)

print("Q1:", q1)
print("Median (Q2):", q2)
print("Q3:", q3)
```

```python
plt.figure()
plt.scatter(df["Age"], df["Spending Score (1-100)"])

plt.title("Age vs Spending Score")
plt.xlabel("Age")
plt.ylabel("Spending Score")
plt.grid(True)

plt.show()
```

```python
sns.pairplot(df, hue="Gender")
plt.show()
```

```python
plt.figure(figsize=(8,6))

corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```

```python
print(corr.unstack().sort_values(ascending=False))
```

```python
from sklearn.preprocessing import MinMaxScaler
# select only real features (exclude ID)
features = ["Age", "Annual Income (k$)", "Spending Score (1-100)"]

scaler = MinMaxScaler()
df_scaled = df.copy()
df_scaled[features] = scaler.fit_transform(df[features])
```

```python
plt.figure()

plt.plot(df["Annual Income (k$)"].values, label="Original")
plt.plot(df_scaled["Annual Income (k$)"].values, label="Scaled")
plt.title("Before vs After Normalization")
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12,5))
# histogram
ax[0].hist(df["Age"], bins=15)
ax[0].set_title("Age Histogram")
# scatter
ax[1].scatter(df["Age"], df["Spending Score (1-100)"])
ax[1].set_title("Age vs Spending Score")

plt.show()
```

```python
plt.figure(facecolor="lightgray")

plt.scatter(df["Age"],
            df["Spending Score (1-100)"],
            marker="x",
            label="Customers")
plt.title("Customized Scatter Plot")
plt.legend()
plt.show()
```
